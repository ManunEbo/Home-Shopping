<h3 style="color: green;">Current dataset with drift</h3>

```python
wk_list =['Nbr_trips_per_wk','Drinks_wk','Expenditure_per_wk','Nbr_items_per_wk',
'Cosmetics_and_selfcare_wk','Raw_meats_wk','Vegetables_wk','Bread_exp_wk',
'Total_Price','Eating_out_wk','Drinks_exp_wk','Dairy_produce_wk',
'Card_Source_num','Cooked_meats_wk','Tech_and_services_exp_wk',
'Total_Nbr_of_Items','Raw_meats_exp_receipt','Week_day_numeric',
'Snacks_wk','Bread_wk','Raw_meats_exp_wk','Vegetables_exp_receipt']

wk_perc_list = ['Total_Exp_wk_perc','Nbr_items_wk_perc',
                'Bread_wk_exp_perc','Dairy_produce_wk_exp_perc','Vegetables_wk_exp_perc']
```

```python
# reading in current_data_no_drift
df = pd.read_csv("../Data/current_data_no_drift.csv")
```

```python
df[wk_list].head()
```

```python
df1 = df
```

```python
# Adding random noise to the non percentage features
# In general all the values are being inflated
mu, sigma = 10,6 
for item in wk_list:
    df1['{}'.format(item)] = df1['{}'.format(item)] + \
    abs(np.random.normal(mu, sigma,len(df1)))
df1[wk_list].head()
```

```python
df[wk_perc_list].head()
```

```python
# Adding random noise to the weekly percentages features
mu, sigma = 0,0.1
for item in wk_perc_list:
    df1['{}'.format(item)] = df1['{}'.format(item)] + \
    abs(np.random.normal(mu, sigma,len(df1)))
df1[wk_perc_list].head()
```
